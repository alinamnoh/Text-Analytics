```python
# import libraries
import re
import json
import requests
from datetime import datetime
```

```python
# define variables that will be used
offset = 0 # to point to the start/end
reviews = [] # a reviews array to store the reviews
```

```python
# get the URL for the product chosen
url = 'https://shopee.com.my/Crossbody-Bags-for-Women-Soft-Leather-Luxury-Handbags-Women-Bags-Designer-Female-Shoulder-Messenger-Bag-Totes-i.238433890.7752093894?sp_atk=50c16a8d-8b14-40ec-8655-3cf5e5aba91a&xptdk=50c16a8d-8b14-40ec-8655-3cf5e5aba91a'

# search particular patterns in URL using regular expression
r = re.search(r'i\.(\d+)\.(\d+)', url)

# conidtional statement if...else to know either shop_id and item_id exists in r to validate if the URL is valid or not
if r:
    shop_id, item_id = r.group(1), r.group(2)
else:
    print("Invalid URL")
    exit()
```

```python
# fetch data from the URL and parse the JSON response to interact with its contents
# to also validate if the data is fetched successfully or not using conditional statement if...else
ratings_url = f'https://shopee.com.my/api/v2/item/get_ratings?filter=0&flag=1&itemid={item_id}&limit=20&offset={offset}&shopid={shop_id}&type=0'
response = requests.get(ratings_url)

if response.ok:
    data = response.json()
    print("Data fetched successfully") 
else:
    print("Failed to fetch data.")
```

```python
# loop until condition is met
offset = 0
while True:
    # set the limit to only 20 reviews
    ratings_url = f'https://shopee.com.my/api/v2/item/get_ratings?filter=0&flag=1&itemid={item_id}&limit=20&offset={offset}&shopid={shop_id}&type=0'
    data = requests.get(ratings_url).json()
    
    # from the data fetched, for loop will iterate to every data to extract the author_username, ctime (and modified into appropriate date format) and comment
    for rating in data['data']['ratings']:
        reviewer = rating['author_username']
        time = rating['ctime']
        modified_date = datetime.fromtimestamp(time).strftime('%Y-%m-%d')
        content = rating['comment']
    
        # append into reviews list
        reviews.append({
            'REVIEWER': reviewer,
            'DATE': modified_date,
            'CONTENT': content
        })
    # exit the loop if the length of the list (fetched data) is less than 20
    if len(data['data']['ratings']) < 20:
        break
    offset += 20
    
```

```python
# print every data in reviews
for review in reviews:
    print("Reviewer: ", review['REVIEWER'])
    print("Date: ", review['DATE'])
    print("Content: ", review['CONTENT'])
    print()
```

```python
# import csv library
import csv
```

```python
# write the review into csv file named shoppeReviews
with open('shopeeReviews.csv', 'w', newline = '', encoding = 'utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames = ['REVIEWER', 'DATE', 'CONTENT'])
    writer.writeheader()
    writer.writerows(reviews)
    
print("Saved successfully into csv file")
```
